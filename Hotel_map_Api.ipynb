# Creacion Mapa

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv("Data/df_hotels_22&21.csv", sep=',', encoding='UTF-8', index_col=0)
pd.options.display.max_columns = None
df.head()
```

```python
df.head(1)
```

```python
df_map = df[["Hotel", "Location", "Country", "Region"]]
df_map.head(1)
```

```python
def join_columns(row):
    return f"{row['Location']}, {row['Country']}"

df_map["Address"] = df_map.apply(join_columns, axis=1)
```

```python
df_mapprueba= df_map.head(80)
```

tengo que conseguir que todos esten rellenados

```python
geolocator = Nominatim(user_agent="geoapiExercises")

# Agregar nuevas columnas para latitud y longitud
df['latitude'] = None
df['longitude'] = None

# Iterar sobre cada fila en el DataFrame
for index, row in df_mapprueba.iterrows():
    
    # Busca la ubicación de la ciudad y pais
    location = geolocator.geocode(row['Location']+','+row['Country'])
    # Asigno la latitud y longitud a las columnas correspondientes
    if location is not None:
        df.at[index, 'latitude'] = location.latitude
        df.at[index, 'longitude'] = location.longitude
```

```python
df[["latitude", "longitude"]] = df[["latitude", "longitude"]].astype(float)
df_coord = df.head(78)
df_coord = df_coord.fillna(0)
df_coord
```

```python
# crear mapa

mapa=folium.Map(location=[df.latitude.mean(), df.longitude.mean()],  # la posicion
                tiles='openstreetmap',            # tipo de mapa, theme
                zoom_start=2)                    # zoom inicial

mapa
```

```python
for index, row in df_coord.iterrows():# método para iterar por los dataframes
    if row['Rank_Year'] == 2022:
        folium.Marker([row["latitude"], row['longitude']],
                      radius=10, 
                      tooltip = row['Hotel'],
                      icon=folium.Icon(icon = "home", color = "green")).add_to(mapa)
    elif row['Rank_Year'] == 2021:
        folium.Marker([row["latitude"], row['longitude']],
                      radius=10, 
                      tooltip = row['Hotel'],
                      icon=folium.Icon(icon = "info-sign", color = "red")).add_to(mapa)
    else:
        pass
```
